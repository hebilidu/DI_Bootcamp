## Linear Regression Part 1: Adding Noise

**Questions:**
1. Sample 1000 data points (x) from a uniform distribution over $(1,10)$. (Hint: use np.random.uniform.)
2. Sample noise $n$ from $\mathcal{N}(0, 100)$ ($\sigma^2 = 100$) for each data point. Define $y = 5x + 2$ and plot $y=y + n$ with plt.scatter(..., alpha=0.2). (Hint: Use np.random.normal, and be careful not to confuse variance and standard deviation.)
3. Perform linear regression to generate $\hat{y}$, an approximation of $y$, using $x$ and $y'$. Plot $\hat{y}$ on top of the plot from question 2.
4. What is the equation for the regression line (to two decimal places)? (See .coef_ and .intercept_ attributes in LinearRegression.) The result should be close to, but not equal to, the line from question 1.
5. Compute the $R^2$ score using LinearRegression.score().
6. If we instead sampled noise $n$ from $\mathcal{N}(0, 36)$ in question 2, what $R^2$ score would we get? Why?

```python
# 1
np.random.seed(0)
x = np.random.uniform(low=1, high=10, size=1000)
```

```python
# 2
n = np.random.normal(loc=0, scale=10, size=1000)
y = 5*x + 2
y_prime = y + n

plt.scatter(x, y_prime, alpha=0.2)
plt.xlabel('x')
plt.ylabel('y\'')
plt.title('Y on noisy Y')
plt.show()
```

```python
# Approximation of y
x = x.reshape(-1,1)
reg = LinearRegression().fit(x, y_prime)
y_approx = reg.predict(x)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x, y_prime, alpha=0.2)
plt.plot(x,y_approx, color='r', linewidth=3, label='approximated y')
plt.xlabel('x')
plt.ylabel('y\'')
plt.legend()
plt.title('Y on noisy Y')
plt.show()
```

```python
# 4
print("The approximated coefficient is {:.2f}".format(reg.coef_[0]))
print("The approximated intercept is {:.2f}".format(reg.intercept_))
```

```python
# 5
print("The R2 score is {:.2f}".format(reg.score(x,y_prime.reshape(-1,1))))
```

```python
# With a sampled noise from (0,36) normal we would get a higher R2 as the variability of the data from the affine function decreases.
n2 = np.random.normal(loc=0, scale=6, size=1000)
y_prime2 = y + n2
reg2 = LinearRegression().fit(x, y_prime2)
print("The R2 score is {:.2f}".format(reg2.score(x,y_prime2.reshape(-1,1))))
```

Indeed R2 is higher in this case.

# Part 2: Anomalous Data

**Questions:**
  7. Load the data in *lin_reg.csv* to a pandas data frame, and perform a linear regression on the x and y columns. Print the intercept, coefficient, and $R^2$ values.
  8. Plot the residuals for the regression performed on the all rows of the data frame. What do you see?
  9. Plot a scatter plot of the x and y values, with the regression line on top. What do you see?
  10. How would you improve the linear regression? What would be better intercept and coefficient values? Plot another scatter plot of the x and y values with this improved linear regression line on top.

```python
df = pd.read_csv('lin_reg.csv')
```

```python
df.shape # (505,2)
```

```python
x = df['x'].values.reshape(-1,1)
y = df['y'].values.reshape(-1,1)
reg = LinearRegression().fit(x, y)
print("The approximated coefficient is {:.3f}".format(reg.coef_[0][0]))
print("The approximated intercept is {:.3f}".format(reg.intercept_[0]))
print("The R2 score is {:.3f}".format(reg.score(x,y)))
```

```python
residuals = y - reg.predict(x)
plt.scatter(df.index,residuals)
plt.show()
```

We see that except for a few points that have high residual values, residuals are low.

```python
plt.figure(figsize=(10,5))
plt.scatter('x', 'y', data=df, label='data', alpha=0.4)
plt.plot(x, reg.predict(x), color='r', linewidth=3, label='prediction')
plt.legend()
plt.show()
```

Indeed a few outliers disturb the regression.

```python
# To improve the model we can remove ouliers.
idx_to_drop = df[df['y']>800].index
df = df.drop(idx_to_drop)
```

```python
x2 = df['x'].values.reshape(-1,1)
y2 = df['y'].values.reshape(-1,1)
reg2 = LinearRegression().fit(x2, y2)
print("The approximated coefficient is {:.3f}".format(reg2.coef_[0][0]))
print("The approximated intercept is {:.3f}".format(reg2.intercept_[0]))
print("The R2 score is {:.3f}".format(reg2.score(x2,y2)))
```

```python
plt.figure(figsize=(10,5))
plt.scatter('x', 'y', data=df, label='data', alpha=0.4)
plt.plot(x, reg2.predict(x), color='r', linewidth=3, label='prediction')
plt.legend()
plt.show()
```
