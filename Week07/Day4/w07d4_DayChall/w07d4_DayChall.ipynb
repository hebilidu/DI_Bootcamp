```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# generate random data METHOD 2 : use make_blobs
X, y = make_blobs(n_samples=300, centers=4, n_features=2, cluster_std=0.5, random_state=4)
plt.scatter(X[:,0], X[:,1])
plt.title("Dataset created with make_blobs")
```

```python
# Train scikit-learn’s KMeans model with 4 clusters
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
```

```python
# fit 
kmeans.fit(X)
print('--------------------- KMeans result --------------------')
# Kmeans is defined by a few parameters - Inertia + Centers Cluster
print(kmeans.inertia_)
print(kmeans.cluster_centers_)
```

```python
dataf = pd.DataFrame(X, columns=['x','y'])
dataf['label'] = kmeans.labels_
c = ['b', 'g', 'k', 'yellow']
plt.figure(figsize=(10,5))

for i in range(4):
    data_temp = dataf[dataf['label'] == i]
    plt.scatter(data_temp['x'], data_temp['y'], c=c[i], label = f'Label {i}')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r')
plt.legend()
plt.title('Predicted labels with cluster centers')
plt.show()
```

```python
km2 = KMeans(n_clusters=2, random_state=0)
km3 = KMeans(n_clusters=3, random_state=0)
km4 = KMeans(n_clusters=4, random_state=0)
km5 = KMeans(n_clusters=5, random_state=0)
```

```python
# fit
c = ['b', 'g', 'k', 'yellow', 'brown']
X2 = X
km2.fit(X2)
X2['label'] = km2.labels_
plt.figure(figsize=(10,5))
for i in range(2):
    data_temp = X2[X2['label'] == i]
    plt.scatter(data_temp['petal_length'], data_temp['petal_width'], c=c[i], label = f'Label {i}')
plt.scatter(km2.cluster_centers_[:,0], km2.cluster_centers_[:,1], c='r')
plt.legend()
plt.title('Predicted labels with 2 cluster centers')
```
